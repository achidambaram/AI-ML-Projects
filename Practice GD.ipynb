```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("/content/housing.csv")

x = dataset['median_income']
y = dataset['median_house_value']

xArray = x.to_numpy()
yArray = y.to_numpy()

print(xArray)
print(yArray)

def gradient_descent(x, y):
  m = 0
  c = 0
  n = len(x)
  iterations = 1000
  learning_rate = 0.001

  for i in range(iterations):
    y_predicted = m * x + c
    cost = (1/n) * sum([val**2 for val in y - y_predicted])
    mS = -(2/n) * sum(x * (y - y_predicted))
    bS = -(2/n) * sum((y - y_predicted))
    m = m - mS * learning_rate
    c = c - bS * learning_rate
    plt.plot(x, m * x + c)
    print("m {}, c {}, cost {}, iterations {}".format(m, c, cost, iterations))
    iterations -= 1

a = np.array([1, 2, 3, 4, 6])
b = np.array([10, 20, 30, 20, 50])

plt.scatter(xArray, yArray)

gradient_descent(xArray, yArray)


  
```
