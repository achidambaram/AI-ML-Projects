```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans 
import cv2
```

```python
image = cv2.imread('/content/redBackground.jpeg')

def imgtoRGBArray(image):
  alteredImg = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  plt.imshow(alteredImg)

  print(alteredImg)

  print("Length: {} Width: {} Height: {}".format(len(alteredImg), len(alteredImg[0]), 3))

  return alteredImg

def RGB2HEX(imgArray):
  hexArray = "#{:02x}{:02x}{:02x}".format(int(imgArray[0]), int(imgArray[1]), int(imgArray[2]))
  return hexArray

alteredImg = imgtoRGBArray(image)
hexImgArray = RGB2HEX(alteredImg)
print(hexImgArray)


        
```
