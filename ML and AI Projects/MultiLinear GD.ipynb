```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("/content/cars.csv", ";")
dataset.drop([0],  inplace=True)
dataset.head()

x1 = dataset['MPG']
x2 = dataset['Horsepower']
y = dataset['Acceleration']

x1Array = x1.to_numpy().astype(float)
x2Array = x2.to_numpy().astype(float)
yArray = y.to_numpy().astype(float)

print(np.shape(x1Array))

def gradient_descent(x1, x2, y):
  m1_curr = m2_curr = b_curr = 0
  n = len(x1)
  iterations = 100
  learning_rate = 0.03

  for i in range(iterations):
    y_predicted = m1_curr * x1 + m2_curr * x2 + b_curr
    cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
    m1d = (-2/n) * sum(x1 * (y - y_predicted)) 
    m2d = (-2/n) * sum(x2 * (y - y_predicted)) 
    bd = (-2/n) * sum(y - y_predicted) 

    m1_curr = m1_curr - (learning_rate * m1d)
    m2_curr = m2_curr - (learning_rate * m2d)
    b_curr = b_curr - (learning_rate * bd)

    print("m1 {}, m2 {}, b {}, cost {}, iterations {}".format(m1_curr, m2_curr, b_curr, cost, i))
    plt.plot(x1, x2, m1_curr * x1 + m2_curr * x2 + b_curr)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
plt.scatter(x1Array, x2Array, yArray)

gradient_descent(x1Array, x2Array, yArray)

```
